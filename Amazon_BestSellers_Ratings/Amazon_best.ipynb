```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service
from webdriver_manager.firefox import GeckoDriverManager
import time
import random

firefox_options = Options()
#firefox_options.add_argument('--headless')
firefox_options.add_argument('--no-sandbox')
firefox_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)

time.sleep(3)
driver.quit()

driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)

driver.get(url) 
    
#assert 'Zara' in driver.title 
    
SCROLL_PAUSE_TIME = 20

   
last_height = driver.execute_script("return document.body.scrollHeight")
wait = WebDriverWait(driver, 10)
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    wait = WebDriverWait(driver, 20)
wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "[href='/gp/help/customer/display.html?nodeId=468496&ref_=footer_privacy']")))
  
html = driver.page_source 
soup = bs(html, 'html.parser')
```

```python
name=[]
price=[]
color=[]
ratings_c=[]
ratings=[]
info=[]
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service
from webdriver_manager.firefox import GeckoDriverManager
import time
import random

firefox_options = Options()
#firefox_options.add_argument('--headless')
firefox_options.add_argument('--no-sandbox')
firefox_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)

time.sleep(3)
driver.quit()


driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)


for i in range(len(links)-25):
    
    driver.get(links[i])
    html1 = driver.page_source 
    soup1 = bs(html1, 'html.parser')
    
    name.append(soup1.find_all('div',id="titleSection")[0].get_text().strip())
    price.append(soup1.find_all('span',class_="a-price-whole")[0].get_text().replace(".",""))
    color.append(soup1.find_all('div',id='variation_color_name')[0].find_all('div')[0].find_all('span')[0].get_text().strip())
    ratings_c.append(soup1.find_all('div',id="averageCustomerReviews")[0].find_all('span',id="acrCustomerReviewText")[0].get_text())
    ratings.append(soup1.find_all('div',id="averageCustomerReviews")[0].find_all('i')[0].get_text())
    a=soup1.find_all('li',class_='a-spacing-mini')
    inf=[]
    for i in a:
        inf.append(i.get_text())
    info.append("".join(inf))
    


```

```python
Output_File = open("C:/Users/oyo/Desktop/Web data/Amazon_best.csv", "w", encoding="utf-8")

Output_File.write("Name,Color,Price,Info,Ratings_count,Ratings\n")
for x in range(0, len(name)):
    Output_File.write(str(name[x].replace(','," ")) + ',' + str(color[x].replace(','," ")) + ','+ str(price[x].replace(','," ")) + ','+str(info[x].replace(','," ")) + ','+str(ratings_c[x].replace(','," ")) +','+str(ratings[x].replace(','," ")) +'\n')

Output_File.close()
```
