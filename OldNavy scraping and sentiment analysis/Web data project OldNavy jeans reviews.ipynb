```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service
from webdriver_manager.firefox import GeckoDriverManager
import time
import random

firefox_options = Options()
#firefox_options.add_argument('--headless')
firefox_options.add_argument('--no-sandbox')
firefox_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)

time.sleep(3)
driver.quit()


driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)
driver.get(url) 
    

    
SCROLL_PAUSE_TIME = 20

   
# Get the initial height of the webpage
last_height = driver.execute_script("return document.body.scrollHeight")

# Scroll down the webpage until reaching the bottom
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
# If the page height hasn't increased, break out of the loop    if new_height == last_height:
        break
    last_height = new_height
    wait = WebDriverWait(driver, 10)
wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "[href='https://www.gapinc.com/en-us/consumer-privacy-policy']")))
  
html = driver.page_source 
soup = bs(html, 'html.parser')

relevant = soup.find_all('div', class_='cat_product-image category-page-mvez9e')

product_links=[]
for li in relevant:
# Find all <a> elements with class 'category-page-0' within the <div>
    for html_class in li.find_all('a', class_="category-page-0"):
# Get the 'href' attribute (the product link) and append it to the list
        link=html_class.get('href')
        product_links.append(link)
        
pid=[]
# Extract product IDs from the collected links
for l in product_links:
    st=l.find("pid=")
    end=l.find('&cid')
 # Extract the product ID substring and append it to the list
    pid.append(l[st+4:end])
    
pid[0:100]
```
