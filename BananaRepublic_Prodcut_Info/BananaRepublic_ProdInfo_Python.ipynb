```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service
from webdriver_manager.firefox import GeckoDriverManager
import time
import random

firefox_options = Options()
#firefox_options.add_argument('--headless')
firefox_options.add_argument('--no-sandbox')
firefox_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)

time.sleep(3)
driver.quit()


driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)
driver.get(url) 
    
#assert 'Zara' in driver.title 
    
SCROLL_PAUSE_TIME = 10

   
last_height = driver.execute_script("return document.body.scrollHeight")
wait = WebDriverWait(driver, 10)
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    wait = WebDriverWait(driver, 20)
wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "[href='https://www.gapinc.com/en-us/consumer-privacy-policy']")))
  
html = driver.page_source 
soup = bs(html, 'html.parser')
```

```python
name=[]
color=[]
price=[]
material=[]
info=[]
firefox_options = Options()
#firefox_options.add_argument('--headless')
firefox_options.add_argument('--no-sandbox')
firefox_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)

time.sleep(3)
driver.quit()


driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)

for i in links:
    driver.get(i)

    html1 = driver.page_source 
    soup1 = bs(html1, 'html.parser')
    
    name.append(soup1.find_all('div', id="product")[0].find_all('h1')[0].get_text())
    color.append(soup1.find_all('div',class_='swatch-label pdp-mfe-j0t5em')[0].get_text())
    price.append(soup1.find_all('div',class_='amount-price')[0].get_text())
    material.append(soup1.find_all('div', class_="product-information-item__information pdp-mfe-b68baj")[0].find_all('div')[3].get_text())
    info.append(soup1.find_all('div', class_="product-information-item__information pdp-mfe-b68baj")[0].find_all('div')[2].get_text())
```

```python
Output_File = open("C:/Users/oyo/Desktop/Web data/BR_new.csv", "w", encoding="utf-8")

Output_File.write("Name,Color,Price,Info,Material\n")
for x in range(0, len(name)):
    Output_File.write(str(name[x].replace(','," ")) + ',' + str(color[x].replace(','," ")) + ','+ str(price[x].replace(','," ")) + ','+str(material[x].replace(','," ")) + ','+str(info[x].replace(','," ")) + '\n')

Output_File.close()
```
